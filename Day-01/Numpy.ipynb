```python
#CREATE ARRAY

import numpy as np

a = np.array([1, 2, 3, 4])

print(a)
print(type(a))
```

```python
#MATRIX

import numpy as np

matrix = np.array([
    [1,2,3],
    [4,5,6]
])

print(matrix)
```

```python
#SHAPE
print(matrix.shape)
```

```python
#ADDITION
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a+b)
```

```python
#MULTIPLICATION
print(a*b)
```

```python
#DOT PRODUCT
print(np.dot(a,b))
```

```python
#STATISTICS
data = np.array([10,20,30,40,50])

print(np.mean(data))
print(np.std(data))
print(np.max(data))
print(np.min(data))
```
